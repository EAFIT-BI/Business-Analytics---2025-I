# Clustering Jerárquico

Tiene dos vertientes:

1. Arriba-abajo : parte del principo de que todas las observaciones son un solo grupo. Después, de forma iterativa empieza a realizar divisiones con base en la distancia intra-cluster (se desea que sea muy pequeña)
2. Abajo-arriba: parte de que cada registro u observación es un cluster y, de forma iterativa, los agrupa hasta que la distancia inter-cluster sea la máxima posible

La forma en la que se selecciona el mejor número de clusters es con el coeficiente de **silueta**, que también se usa para la evaluación de los algoritmos de agrupamiento

```python
# Importamos librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generamos el toyset
from sklearn.datasets import make_blobs

X,y  = make_blobs(n_samples=1000,
                  n_features=2, centers=4,
                  cluster_std=0.6,
                  shuffle=True, random_state=0)

# Graficamos el conjunto
plt.scatter(X[:,0], X[:,1], s = 50)
plt.show()
```

```python
# Vamos a entrenar nuestro modelo fijando el número de
# clusters
from sklearn.cluster import AgglomerativeClustering

# Instanciamos el modelo de clustering
model = AgglomerativeClustering(n_clusters = 2)
# Entrenamos el modelo
model.fit(X)
# Encontramos el y_pred, es decir, el cluster al que pertenece
# cada observación
```

```python
y_pred = model.fit_predict(X) # aquí se generan los clusters
```

```python
# Graficamos el dendograma, que muestra la organización
# de las jerarquías para la construcción de los cluster

from scipy.cluster.hierarchy import dendrogram, linkage

Z = linkage(X, method='ward')
plt.figure(figsize=(25, 10))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('sample index')
plt.ylabel('distance')
dendrogram(Z, leaf_rotation = 0, leaf_font_size = 8,
           orientation = 'right')
plt.show()
```

```python
# Vamos a visualizar los dos grupos en el diagrama de dispersión
y = model.labels_
plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap='jet')
plt.show()
```
