#Transformaciones

Las transformaciones se dan usualmente porque el  formato actual de los datos no es apropiado o suficiente para que sea informativo. Las transformaciones más comunes son:

- Operaciones entre columnas
- Operaciones con escalares
- Cambio de tipo de datos
- Medidas de resumen

```python
#Vamos a comparar los movimientos débito y crédito

#Primero sumamos todos los movimientos del débito
suma_debito = estados_financieros['Movimientos Debito'].sum()

#Segundo sumamos todos los movimientos del crédito
suma_credito = estados_financieros['Movimientos Credito'].sum()

#Finalmente, para comparar haremos un condicional
if suma_debito > suma_credito: #revisamos si el débito es mayor al crédito
  print("El débito es mayor")
elif suma_debito < suma_credito: #revisamos si el débito es menor al crédito
  print("El crédito es mayor")
else: #revisamos si hay una igualdad
  print("Son iguales")
```

```python
#Queremos determinar qué tipo de transacción es más frecuente

#Podemos imprimir el mensaje sin necesidad de almacenar el resultado
#previamente en alguna variable

print("La transacción más frecuente es: ",
      estados_financieros['Descripcion'].mode()[0])

#dado que podemos tener varias modas vamos a seleccionar la primera
#(o sea la de la posición 0)
```

```python
#¿Cuál es el máximo saldo final?¿Cuál es la cuenta que corresponde a este saldo?

#Almacenamos el máximo en una variable llamada maximo_saldo
maximo_saldo = estados_financieros['Saldo Final'].max()

#Recuperamos la cuenta que corresponde a esa valor máximo
cuenta = estados_financieros[estados_financieros['Saldo Final'] == maximo_saldo]

#Imprimimos la respuesta de la solicitud
print("El máximo saldo final es de: $", maximo_saldo,
      "y la cuenta a la que pertenece es: ", cuenta['Descripcion'])
```

```python
#Nos solicitan encontrar la variación de los saldos

estados_financieros['Variación'] = estados_financieros['Saldo Final'] - \
estados_financieros['Saldo Inicial']

#mostramos la tabla resultante con la nueva  columna (los 3 últimos registros)
estados_financieros.tail(3)
```

```python
#Mostrar el promedio de los saldos para cada  tipo de cuenta

#1. Se seleccionan las columnas a mostrar
#2. Se decide la columna que agrupa
#3. Se aplica la medida de resumen

#1. Descripcion, Saldo Inicial, Saldo Final
#2. Descripcion
#3. Media

#Los dobles corchetes se ponen debido a que se selecciona
#más de una columna
estados_financieros[['Descripcion', 'Saldo Inicial', 'Saldo Final']].\
groupby('Descripcion').mean()
```

#Gráficos

```python
#Graficar el código vs los movimientos
estados_financieros.plot(x = 'Codigo', y = ['Movimientos Credito', 'Movimientos Debito'])
plt.title("Movimientos por código") #título del gráfico
plt.xlabel("Código") #nombre del eje x
plt.ylabel("Movimientos") #nombre del eje y
plt.grid() #malla del plano
plt.show()
```

```python
#Revisamos los tipos de cuenta que tiene la tabla
estados_financieros['Descripcion'].unique()
```

```python
#De todas las cuentas graficaremos el saldo final de ACTIVOS y CAJA
#Hacemos el filtrado

#selecciona todos los registros que pertenecen a las cuentas CAJA o ACTIVOS
datos = estados_financieros[estados_financieros['Descripcion'].isin(['Licencias'])]
```

```python
#Graficamos
sns.boxplot(x = 'Descripcion', y = 'Saldo Final', data = datos)
plt.title("Saldo Final por Cuenta")
plt.xlabel("Cuenta")
plt.ylabel("Saldo Final")
plt.show()
#
```
